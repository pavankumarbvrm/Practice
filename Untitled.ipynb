```python
import mimetypes
from datetime import datetime, timedelta

from boto.s3.connection import S3Connection


AWS_KEY = 'AKIAIRAHDENDH3ZBXEIQ'
AWS_SECRET = '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'
AWS_BUCKET_NAME = 'webinartesting.shawacademy.com'
PREFIX = 'images'


def main():
    conn = S3Connection(AWS_KEY, AWS_SECRET)
    bucket = conn.get_bucket(AWS_BUCKET_NAME)
    key_list = bucket.get_all_keys(prefix=PREFIX)
    for key in key_list:
        content_type, unused = mimetypes.guess_type(key.name)
        if not content_type:
            content_type = 'text/plain'
        expires = datetime.utcnow() + timedelta(days=(7))
        expires = expires.strftime("%a, %d %b %Y %H:%M:%S GMT")
        metadata = {'Expires': expires, 'Content-Type': content_type}
        print (key.name, metadata)
        key.copy(AWS_BUCKET_NAME, key, metadata=metadata, preserve_acl=True)


if __name__ == '__main__':
    main()

```
